## Deliverable 2: Scrape and Analyze Mars Weather Data

```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup 
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
#Set up Splinter
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = 'https://static.bc-edx.com/data/web/mars_facts/temperature.html'
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
#Scrape the website
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Create a Beautiful Soup Object
table = soup.find_all('tr',class_='data-row')
table
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
#Create a list and fill with row data from the table
mars_temp = []
for t in table:
    row_heading = t.find_all('td')
    row_data = [col.text for col in row_heading]
    mars_temp.append(row_data)    
mars_temp   
```

```python
# Confirm DataFrame was created successfully
mars_df = pd.DataFrame(mars_temp, columns=['id', 'terrestrial_date','sol', 'ls','month','min_temp','pressure'])
mars_df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
mars_df.dtypes
```

```python
# Change data types for data analysis
mars_df = mars_df.astype({'id':int,
                          'sol':int,
                          'ls':int, 
                          'month':int,
                          'min_temp':float, 
                          'pressure':float})
mars_df['terrestrial_date']= pd.to_datetime(mars_df['terrestrial_date'])
```

```python
# Confirm type changes were successful by examining data types again
mars_df.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
months = mars_df['month'].max() 
months
print(f'There are {months} months on Mars')
```

```python
# 2. How many Martian days' worth of data are there?
sol_min = mars_df['sol'].min()
sol_max = mars_df['sol'].max()
martian_days = sol_max - sol_min
martian_days
print(f'There are {martian_days} martian days worth of data')
```

```python
#3.What is the average low temperature by month?
avg_min_temp = mars_df.groupby('month')['min_temp'].mean()
avg_low_temp = avg_min_temp.to_frame()
avg_low_temp
```

```python
#Find Coldest and Warmest months in a year on Mars
df=avg_low_temp.sort_values('min_temp')
df1 = df.reset_index()
coldest = df1.loc[0,'month']
print(f'The coldest month on Mars is month {coldest}')
df2 = df1.sort_values('min_temp', ascending=False)
df3 = df2.reset_index()
warmest = df3.loc[0,'month']
print(f'The warmest month on Mars is month {warmest}')
```

```python
# Plot the average temperature by month
df.plot(kind ='bar')
plt.ylabel("temperature")
plt.tight_layout()
```

```python
# 4. Average pressure by Martian month
avg_pressure = mars_df.groupby('month')['pressure'].mean()
avg_pressure = avg_pressure.to_frame()
avg_pressure
```

```python
#Finding months with lowest and highest pressure
press = avg_pressure.sort_values('pressure')
press1 = press.reset_index()
low_press = press1.loc[0,'month']
low_press
print(f'The month with the lowest average pressure is  month {low_press}')
press2 = press1.sort_values('pressure', ascending=False)
press3 = press2.reset_index()
high_press = press3.loc[0,'month']
high_press
print(f'The month with the highest average pressure is month {high_press}')
```

```python
# Plot the average pressure by month
press.plot(kind='bar')
plt.ylabel("pressure")
plt.tight_layout()
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?
#Locating the first solar longitude
first_ls = mars_df['ls'].iloc[0]
first_ls
```

```python
# Locating the second occurence of the ls (155)
recur_ls = mars_df.loc[mars_df['ls']== first_ls]
recur_ls
```

```python
#Find number of days from the first time ls was 155 and the second time(1 martian year/once around the sun) 
first_ls = recur_ls['terrestrial_date'].iloc[0]
second_ls = recur_ls['terrestrial_date'].iloc[1]
mars_year = second_ls - first_ls
print(f'There are {mars_year} earth days in a mars year')
```

```python
daily_temp = mars_df[['min_temp']]
plt.rcParams["figure.figsize"] = (15,8)
daily_temp.plot()
plt.xticks(np.arange(0,len(mars_df),100))
plt.xlabel('Number of Days')
plt.ylabel('Minimum Temperature')
plt.show()
```

# The above graph shows the distance between the peaks(1425 - 740)to be approximately 685 days.
